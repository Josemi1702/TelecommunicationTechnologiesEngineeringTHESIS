```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler


```

# Ruido con una distribuci√≥n Uniforme

```python


X,Y= make_blobs(n_samples=200, centers=2, n_features=2, cluster_std=4, random_state=42)
random_noise=np.random.uniform(low=-20, high=20, size=(200,2))
X+=random_noise
plt.scatter(X[:,0], X[:,1],c=Y)
```

```python
df= pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"]= Y
df.to_csv("Uniform_Noise.csv")
df
```

# Ruido con una distribuci√≥n Gaussiana

```python

X, Y = make_blobs(n_samples=200, centers=2, n_features=2, cluster_std=4, random_state=42)
random_noise = np.random.normal(loc=0, scale=18, size=(200, 2)) #media 0 y desviaci√≥n standard 18
X += random_noise
plt.scatter(X[:,0], X[:,1], c=Y)
```

```python
df= pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"]= Y
df.to_csv("Gauss_Noise.csv")
df
```

# Ruido exponencial

```python
X, Y = make_blobs(n_samples=200, centers=2, n_features=2, cluster_std=4, random_state=42)
random_noise = np.random.exponential(scale=14, size=(200, 2))#Con tasa media de 1/14
X += random_noise
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
df= pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"]= Y
df.to_csv("Exponential_Noise.csv")
df
```

# Ruido gamma

```python
X, Y = make_blobs(n_samples=200, centers=2, n_features=2, cluster_std=4, random_state=42)
random_noise = np.random.gamma(35, 3, size=(200, 2)) #Forma 10 y escala 1
X += random_noise

plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
df= pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"]= Y
df.to_csv("Gamma_Noise.csv")
df
```
