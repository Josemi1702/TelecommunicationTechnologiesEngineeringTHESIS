```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import make_moons, make_circles, make_blobs



```

# Moons

```python
X, Y = make_moons(n_samples=200, noise=0.05, random_state=170)
plt.scatter(X[:,0], X[:,1],c=Y)
```

```python
df= pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"]= Y
df.to_csv("Moons.csv")
df
```

# Circles

```python
X, Y = make_circles(n_samples=200, factor=0.5, noise=0.05, random_state=170)
plt.scatter(X[:,0], X[:,1],c=Y)
```

```python
df= pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"]= Y
df.to_csv("Circles.csv")
df
```

# Anisotropic

```python
# Anisotropicly distributed data
random_state = 170
X, y = make_blobs(n_samples=200, random_state=170)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
X, Y = (X_aniso, y)


plt.scatter(X[:,0], X[:,1],c=Y)
```

```python
df= pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"]= Y
df.to_csv("Anisotropic.csv")
df
```
